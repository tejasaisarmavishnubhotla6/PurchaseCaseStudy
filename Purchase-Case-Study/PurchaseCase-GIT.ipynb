T-test

```
from scipy.stats import ttest_1samp
```

```
t_stat,P_value=ttest_1samp(sample["Purchase"],10000)
```

```
if P_value<0.05:
  print("Reject Null Hypothesis")
else:
  print("Accept Null Hypothesis")
```

```
data_men=data.loc[(data["Gender"]==1)&(data["Age"]==1)]
data_women=data.loc[(data["Gender"]==0)&(data["Age"]==1)]
```

```
data_men.shape
```

```
data_women.shape
```

```
data_men_Sample=data_men.sample(500,random_state=233)
data_women_Sample=data_women.sample(500,random_state=789)
```

```
data_men_Sample.shape
```

```
data_women_Sample.shape
```

```
print(data_men_Sample.Purchase.var())
print(data_women_Sample.Purchase.var())
```

T-test for two sample when the variance is unequal

```
from scipy.stats import ttest_ind
```

```
t_stat2,p_value2=ttest_ind(data_men_Sample["Purchase"],data_women_Sample["Purchase"],equal_var=True)
print(p_value2)
```

```
if p_value2<0.05:
  print("Reject Null Hypothesis")
else:
  print("Accept Null Hypothesis")
```
